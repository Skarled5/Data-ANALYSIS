```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
```

```python
#  plt.show diyerek aslında grafikerli gösterebiliyoruz ama her zaman kullanıcaz o yüzden hep aktif hale getiriyor inline
```

```python
yasListesi = [10,20,30,30,30,40,50,60,70,75]
```

```python
kiloListesi = [20,60,80,85,86,87,70,90,95,90]
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
numpyKiloListesi
```

```python
numpyYasListesi
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"g")
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("Kilo'nun Yaşa Göre Değişiminin Grafiği")
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1**3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g") #(x ekseni,y ekseni,renk)
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"*-b")
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"--r") 
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"r*-") # subplot(m,n,p) m: satır n: sütün p:pencere numarası
plt.subplot(1,2,2)
plt.plot(numpyDizisi1,numpyDizisi2,"b--")
```

```python
benimFigure = plt.figure()
```

```python
figureAxes = benimFigure.add_axes([0.8,0.8,0.3,0.3]) 
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("x ekseni veri ismi")
figureAxes.set_ylabel("y ekseni veri ismi")
figureAxes.set_title("Grafik Başlığı")
```

# add_axes([a,b,c,d]) c:genişlik d:uzunluk a:y ekseninin başlangıc noksası b: x ekseninin başlangıç noktası
 
                                                      

```python
benimFigure
```

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2 = figure2.add_axes([0.2,0.3,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("x ekseni")
eksen1.set_ylabel("y ekseni")
eksen1.set_title("ANA GRAFİK BAŞLIĞI")

eksen2.plot(numpyDizisi2,numpyDizisi1,"b")
eksen2.set_xlabel("x ekseni")
eksen2.set_ylabel("y ekseni")
eksen2.set_title("Küçük Grafik Başlık")
```

```python
(benimFigure,benimEksenler) = plt.subplots(nrows=1,ncols=2)
type(benimEksenler)

for eksen in benimEksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2,"g")
    eksen.set_xlabel("x ekseni")
plt.tight_layout() # arasındaki mesafeyi düzeltir
```

```python
yeniFigur = plt.figure()
# dpi=150

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 2,label = "numpyDizisi ** 2")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 3,label= "numpyDizisi ** 3")
yeniEksen.legend(loc=2) # legend grafikteki sol üsteki yazıyı açığa çıkarır loc ise lokasyondan gelir yer belirtir
```
